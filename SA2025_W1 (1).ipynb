### 1. What is name of car that has the highest horsepower?  

```python

# Reset index safely without inserting old index as a new column
df = df.reset_index(drop=True)

# Define the correct column names based on your provided list
horsepower_column = 'horsepower'
car_name_column = 'name'

# Data Cleaning and Type Conversion
# Convert 'horsepower' column to numeric, coercing non-numeric values to NaN
df[horsepower_column] = pd.to_numeric(df[horsepower_column], errors='coerce')

# Drop rows where 'horsepower' is NaN (i.e., rows with invalid horsepower data)
df.dropna(subset=[horsepower_column], inplace=True)

# Find the car with the highest horsepower
if df.empty:
    print("No valid horsepower data found after cleaning.")
else:
    # Find the row with the maximum horsepower
    max_hp_row = df.loc[df[horsepower_column].idxmax()]

    # Extract the car name and its horsepower value using the confirmed column names
    car_name = max_hp_row[car_name_column]
    horsepower_value = max_hp_row[horsepower_column]

    # Print the result
    print(f"The car with the highest horsepower ({horsepower_value} hp) is: {car_name}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

most_common_origin = filtered_cars['origin'].mode()[0]

print("Most common origin for cars with horsepower > 100 and weight < 3000:", most_common_origin)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Ensure 'horsepower' is numeric
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Drop rows with missing horsepower or weight
df_clean = df.dropna(subset=['horsepower', 'weight'])

# Compute median mpg
median_mpg = df_clean['mpg'].median()

# Filter cars with above-median mpg
above_median_mpg = df_clean[df_clean['mpg'] > median_mpg]

# Compute horsepower-to-weight ratio
above_median_mpg['hp_to_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Find the maximum ratio
max_ratio = above_median_mpg['hp_to_weight'].max()

# Find car(s) with that ratio
best_cars = above_median_mpg[above_median_mpg['hp_to_weight'] == max_ratio]

# Display result
print("Car(s) with best horsepower-to-weight ratio among cars with above-median mpg:")
print(best_cars[['name', 'mpg', 'horsepower', 'weight', 'hp_to_weight']])

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Capitalize origin names for a cleaner legend (optional)
df['origin'] = df['origin'].str.capitalize()

# Drop rows with missing values in the required columns
df_clean = df.dropna(subset=['mpg', 'model_year', 'origin'])

# Group by model year and origin
grouped = df_clean.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=grouped, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Clean dataframe: drop rows with missing values in required columns
df_clean = df.dropna(subset=['name', 'mpg', 'model_year'])

# Group by car name
grouped = df_clean.groupby('name')

# Filter cars appearing in more than one model_year
multi_year_models = grouped.filter(lambda x: x['model_year'].nunique() > 1)

# Now group by name again and compute stats
consistency_stats = (
    multi_year_models
    .groupby('name')
    .agg(
        count=('mpg', 'count'),
        model_years=('model_year', 'nunique'),
        std_mpg=('mpg', 'std'),
        avg_mpg=('mpg', 'mean')
    )
)

# Filter to cars with low variation in mpg
consistent_models = consistency_stats[consistency_stats['std_mpg'] < 1.0]

# Bonus: Sort by count desc, then avg_mpg desc
consistent_models_sorted = consistent_models.sort_values(by=['count', 'avg_mpg'], ascending=[False, False])

# Reset index for display
consistent_models_sorted = consistent_models_sorted.reset_index()

# Display the result
print(consistent_models_sorted[['name', 'count', 'avg_mpg']])
```
